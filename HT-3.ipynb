Задание 1

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
X = pd.DataFrame(boston["data"], columns=boston["feature_names"])

X.head()
```

```python
y = pd.DataFrame(boston["target"], columns=["price"])

y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)

y_pred.shape
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators = 1000, max_depth = 12, random_state = 42)
```

```python
regressor.fit(X_train, y_train.values[:, 0])
```

```python
y_pred = regressor.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

Метод машинного обучения "Дерево решений" дает ручший результат по сравнению с методом линейной регрессии, т.к. в случае дерева R2=0.875 более приближен к 1
